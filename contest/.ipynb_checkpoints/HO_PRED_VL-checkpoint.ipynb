```python

import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python


plotly.tools.set_credentials_file(username='s654231', api_key='M5QbTaUkVZIgxWhG2og6')
dataset = pd.read_csv('LD2_QT201601~201712.csv')

te = -1

DATEd = dataset['DATE'][1:te] + " " + dataset['TIME'][1:te] 

# Create and style traces
q1s = go.Scatter(
    x = list(DATEd),
    y = list(dataset['q1s'][1:te]),
    name = 'q1 solor radation',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 2),
    opacity = 0.8
)
q2s = go.Scatter(
    x = list(DATEd),
    y = list(dataset['q2s'][1:te]),
    name = 'q2 solor radation',
    line = dict(
        color = ('rgb(22, 96, 167)'),
        width = 2),
    opacity = 0.8
)
q3s = go.Scatter(
    x = list(DATEd),
    y = list(dataset['q3s'][1:te]),
    name = 'q3 solor radation',
    line = dict(
        color = '#851e52',
        width = 2),
    opacity = 0.8
)
q4s = go.Scatter(
    x = list(DATEd),
    y = list(dataset['q4s'][1:te]),
    name = 'q4 solor radation',
    line = dict(
        color = '#d3560e',
        width = 2),
    opacity = 0.8
)
data = [q1s, q2s,q3s,q4s]

# Edit the layout
layout = dict(
    title='Solor radtion [kW/m2],',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(count=1,
                    label='YTD',
                    step='year',
                    stepmode='todate'),
                dict(count=1,
                    label='1y',
                    step='year',
                    stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),

    )
)


fig = dict(data=data , layout=layout)
py.iplot(fig)
```

```python
dataset_output = pd.read_csv('LD1_P201601~201712.csv')

S1_O = go.Scatter(
    x = list(DATEd),
    y = list(dataset_output["S1[MW]"][1:te]),
    name = 'S1 [MW]',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 2),
    opacity = 0.8
)
S2_O = go.Scatter(
    x = list(DATEd),
    y = list(dataset_output["S2[MW]"][1:te]),
    name = 'S2 [MW]',
    line = dict(
        color = ('rgb(22, 96, 167)'),
        width = 2),
    opacity = 0.8
)


data = [S1_O,S2_O]

# Edit the layout
layout = dict(
    title=' Total solar power output,',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),

    )
)


fig = dict(data=data , layout=layout)
py.iplot(fig)
```

```python

S1_out_normal =  (dataset_output["S1[MW]"][1:te] - dataset_output["S1[MW]"][1:te].mean()) / (dataset_output["S1[MW]"][1:te].max() - dataset_output["S1[MW]"][1:te].min())
S2_out_normal =  (dataset_output["S2[MW]"][1:te] - dataset_output["S2[MW]"][1:te].mean()) / (dataset_output["S2[MW]"][1:te].max() - dataset_output["S2[MW]"][1:te].min())


q1_s_normal = dataset["q1s"][1:te].astype('float64')
q2_s_normal = dataset['q2s'][1:te].astype('float64')
q3_s_normal =  dataset['q3s'][1:te].astype('float64')
q4_s_normal =  dataset['q4s'][1:te].astype('float64')

q1_s_normal =  (q1_s_normal- q1_s_normal.mean()) / (q1_s_normal.max() - q1_s_normal.min())
q2_s_normal =  (q2_s_normal- q2_s_normal.mean()) / (q2_s_normal.max() - q2_s_normal.min())
q3_s_normal =  (q3_s_normal- q3_s_normal.mean()) / (q3_s_normal.max() - q3_s_normal.min())
q4_s_normal =  (q4_s_normal- q4_s_normal.mean()) / (q4_s_normal.max() - q4_s_normal.min())


q1_t_normal =  dataset['q1t'][1:te].astype('float64')  
q2_t_normal = dataset['q2t'][1:te].astype('float64') 
q3_t_normal = dataset['q3t'][1:te].astype('float64')  
q4_t_normal =  dataset['q4t'][1:te].astype('float64') 

q1_t_normal =  (q1_t_normal- q1_t_normal.mean()) / (q1_t_normal.max() - q1_t_normal.min())
q2_t_normal =  (q2_t_normal- q2_t_normal.mean()) / (q2_t_normal.max() - q2_t_normal.min())
q3_t_normal =  (q3_t_normal- q3_t_normal.mean()) / (q3_t_normal.max() - q3_t_normal.min())
q4_t_normal =  (q4_t_normal- q4_t_normal.mean()) / (q4_t_normal.max() - q4_t_normal.min())


S1_O = go.Scatter(
    x = list(DATEd),
    y = list(S1_out_normal),
    name = 'S1 [MW]',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 2),
    opacity = 0.8
            
)
    
q1s = go.Scatter(
    x = list(DATEd),
    y = list(q1_s_normal),
    name = 'q1 solor radation',
    line = dict(
        color = ('#8CE600'),
        width = 2),
    opacity = 0.8
)
q2s = go.Scatter(
    x = list(DATEd),
    y = list(q2_s_normal),
    name = 'q2 solor radation',
    line = dict(
        color = ('rgb(22, 96, 167)'),
        width = 2),
    opacity = 0.8
)
q1t = go.Scatter(
    x = list(DATEd),
    y = list(q1_t_normal),
    name = 'q1 temperature',
    line = dict(
        color = '#851e52',
        width = 2),
    opacity = 0.8
)
q2t = go.Scatter(
    x = list(DATEd),
    y = list(q2_t_normal),
    name = 'q2 temperature',
    line = dict(
        color='#d3560e',
        width = 2),
    opacity = 0.8
)


data = [S1_O,q1t,q1s,q2t,q2s]

# Edit the layout
layout = dict(
    title=' Total solar power output,',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),

    )
)


fig = dict(data=data , layout=layout)
py.iplot(fig)



```
